```python
# Vamos a utilizar una API publica que genera usuarios RANDOM


# Importar librerías
import requests
import json
import pandas as pd
import psycopg2
```

```python
# La API devuelve un usuario por cada consulta, por lo que ejecutamos un ciclo para obtener los usuarios deseados
total_users = 50
counter = 0
data = []

while counter < total_users:
  # Hacemos el request a la API
  response = requests.get('https://randomuser.me/api/')
  # Convertimos la data
  data_json = json.loads(response.text)
  data.append(data_json['results'][0])
  counter += 1
```

**Se transforma en un DATAFRAME**

```python
# Convertimos la data a un DataFrame
df = pd.json_normalize(data, sep='_')
df
```

```python
# Verificamos los nombres de los campos
df.columns
```

```python
# Verificamos el tipo de cada campo

df.dtypes
```

```python
# Generamos un CSV

df.to_csv('data_entregable_1.csv', index=False)
```
